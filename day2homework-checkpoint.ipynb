Creating plots using the GeoTIFF files:

```python
# used to transform northings and eastings to lat long:
# copied and pasted from the rasterio website

import rasterio
from rasterio.warp import calculate_default_transform, reproject, Resampling

dst_crs = 'EPSG:4326'

with rasterio.open('BedMachineAntarctica v3 surface Cook Ice Shelf TIFF.tif') as src: 
    transform, width, height = calculate_default_transform(
        src.crs, dst_crs, src.width, src.height, *src.bounds)
    kwargs = src.meta.copy()
    kwargs.update({
        'crs': dst_crs,
        'transform': transform,
        'width': width,
        'height': height
    })

    with rasterio.open('sample2.tif', 'w', **kwargs) as dst:
        for i in range(1, src.count + 1):
            reproject(
                source=rasterio.band(src, i),
                destination=rasterio.band(dst, i),
                src_transform=src.transform,
                src_crs=src.crs,
                dst_transform=transform,
                dst_crs=dst_crs,
                #resampling=Resampling.nearest)
                resampling=Resampling.bilinear)


print(f"Source CRS: {src.crs}")
```

```python

def xyz_geotiff(img): #designed to take a tif file, and output the corresponding x, y and z values from that data
    band = img.GetRasterBand(1) #grabs the data in the first band
    z = np.flipud(band.ReadAsArray()).T #flips and transposes data
    width = img.RasterXSize # retrieves the image width
    height = img.RasterYSize # retrieves the image height
    gt = img.GetGeoTransform() # geo transform is the projection / where that data is in 'space'. It converts pixel coordinates to geographic coordinates
    minx = gt[0]
    miny = gt[3] + width*gt[4] + height*gt[5]
    maxx = gt[0] + width*gt[1] + height*gt[2]
    maxy = gt[3]
    lon = np.linspace(minx,maxx,width)
    lat = np.linspace(miny,maxy,height)
    return lon,lat,z
```

```python
# for the above TIFF file:

# northing is -2030646.5824 and easting is 1158133.4733 in EPSG 3031. I need to convert it to EPSG 4326 in order to convert it from polar stereographic to latitude and longitude coordinates.

from pyproj import Transformer

# Define transformation from EPSG:3031 (Antarctic Polar Stereographic) to EPSG:4326 (Lat/Lon)
transformer = Transformer.from_crs("EPSG:3031", "EPSG:4326", always_xy=True)

# Convert easting, northing to latitude/longitude
lon, lat = transformer.transform(1158133, -2030646)

print(f"Latitude: {lat}, Longitude: {lon}")
```
